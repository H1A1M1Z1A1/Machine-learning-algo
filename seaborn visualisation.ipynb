```python
import pandas as pd
import numpy as np
import seaborn as sns
df=sns.load_dataset('tips')
df
```

```python
df.corr()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#category features
sns.countplot(y='day',data=df)
```

```python
sns.catplot(x="sex", y="total_bill",

                hue="smoker", col="time",

                data=df, kind="bar",

                height=4, aspect=.7);
```

```python
sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

```python
sns.catplot(x="day", y="total_bill",

                hue="smoker", palette="Set3", col="time",

                data=df, kind="box",

                height=4, aspect=.7);
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",

                    data=df, palette="Set2", split=True,
                    
                    scale="count", inner="stick",

                    scale_hue=False, bw=.2) 
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```
